```python
np.random.seed(19680801)

# make up some data in the open interval (0, 1)
y = np.random.normal(loc=0.5, scale=0.4, size=1000)
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))

# plot with various axes scales
plt.figure()

# linear
plt.subplot(221)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)

# log
plt.subplot(222)
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
plt.grid(True)

# symmetric log
plt.subplot(223)
plt.plot(x, y - y.mean())
plt.yscale('symlog', linthresh=0.01)
plt.title('symlog')
plt.grid(True)

# logit
plt.subplot(224)
plt.plot(x, y)
plt.yscale('logit')
plt.title('logit')
plt.grid(True)
# Adjust the subplot layout, because the logit one may take more space
# than usual, due to y-tick labels like "1 - 10^{-3}"
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25,
                    wspace=0.35)

plt.show()
```

```python
import matplotlib.collections as collections
```

Массив из чисел от 0 до 2 с шагом 0.01:

```python
t = np.arange(0.0, 2, 0.01)
```

Массив, где каждый элемент - это синус соответствующего элемента из предыдущего массива, умноженного на 2 и на Pi:

```python
s1 = np.sin(2*np.pi*t)
```

Создаём коллекцию из горизонтально расположенных прямоугольников:
1. Для каждого соседней пары элементов из t создаётся прямоугольник с нижним значением 0 и верхним 1.
1. После этого они объединяются в один прямоугольник, к которому будет применена маска s1 > 0, которая разделит их и уберёт лишние части.
1. Результирующие прямоугольники будут зелёного цвета с 50% прозрачностью.

```python
collection1 = collections.BrokenBarHCollection.span_where(t, ymin = 0, ymax = 1, where = s1 > 0, facecolor = 'green', alpha = 0.5)
```

Создаём коллекцию из горизонтально расположенных прямоугольников:
1. Для каждого соседней пары элементов из t создаётся прямоугольник с нижним значением -1 и верхним 0.
1. После этого они объединяются в один прямоугольник, к которому будет применена маска s1 < 0, которая разделит их и уберёт лишние части.
1. Результирующие прямоугольники будут зелёного цвета с 50% прозрачностью.

```python
collection2 = collections.BrokenBarHCollection.span_where(t, ymin = -1, ymax = 0, where = s1 < 0, facecolor = 'red', alpha = 0.5)
```

Строим график:

```python
fig, ax = plt.subplots() #Фигура и график.
ax.set_title('using span_where') #Меняем название графика.
ax.plot(t, s1, color='black') #Строим график. Абсцисса - значения t, ордината - значения s1. Цвет линии - чёрный.
ax.axhline(0, color = 'black', lw = 2) #Добавляем горизонтальную линию с ординатой 0 и толщиной 2.
ax.add_collection(collection1) #Добавляем к графику коллекцию #1.
ax.add_collection(collection2) #Добавляем к графику коллекцию #2.
```

```python
import pandas as pd
sns.set_theme(style = "whitegrid") #Меняем стиль графика - добавляем сетку.
rs = np.random.RandomState(365) #Инициализируем генератор случайных чисел.
values = rs.randn(365, 4).cumsum(axis = 0) #Создаём массив из 365 строк и 4 столбцов из случайных чисел нормального распределения. По каждой строке считаем кумулятивную сумму.
dates = pd.date_range("1 1 2016", periods = 365, freq = "D") #Создаём набор из дат, начиная с 1 января 2016 года. Частота - каждый день. Количество - 365.
data = pd.DataFrame(values, dates, columns=["A", "B", "C", "D"]) #Из дат и полученного массива строим датасет. Названия строк - даты, названия столбцов - буквы A, B, C, D.
data = data.rolling(7).mean() #Находим скользящие средние с размером окна 7.
sns.lineplot(data=data, palette="tab10", linewidth=2.5) #Строим график с полученными данными. Палитра - tab10. Ширина линии - 2.5.
```
